```python
# Conectando com a API
import requests
import pandas as pd

url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&interval=5min&apikey=DFFS0TVTN2FMLCL8'
r = requests.get(url)
base = r.json()
```

```python
base_alpha = pd.DataFrame.from_dict(base['Time Series (Daily)'], orient='index')
base_alpha = base_alpha.transpose()
base_alpha
```

```python
base_alpha = base_alpha.T
base_alpha
```

```python
import numpy as np
```

```python
# Target Binario
base_alpha_num = base_alpha.astype(float)
base_alpha_num['target'] = (base_alpha_num['4, close'].shift(-1) > base_alpha_num['4, close'])
base_alpha_num.dropna(inplace=True)
base_alpha_num
```

```python
# Target Numérico
base_alpha_reg = base_alpha.astype(float)
base_alpha_reg['6. target_reg'] = (base_alpha_reg['4, close'].pct_change().shift(-1))
base_alpha_reg
```

```python
# Target de Tendencia (Coluna Auxiliar)
base_alpha_med = base_alpha.astype(float)
base_alpha_med['6. avg'] = (base_alpha_med['4. close'].rolling(window=3).mean().shift(-3))
```

```python
# Target de Tendencia
base_alpha_med['7. trend'] = (base_alpha_med['6. avg'] > base_alpha_med['4. close'].astype(int))
base_alpha_med
```
